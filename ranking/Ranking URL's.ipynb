```python
data['vertical']=data['url'].str.split('/').str[1]
```

```python
data['searchVolume MONTHLY']=data['searchVolume MONTHLY'].replace(['-',np.NaN,np.nan],'0')
data['Transactions']        =data['Transactions'].replace(['-',np.NaN,np.nan],'0')
data['SC Clicks']           =data['SC Clicks'].replace(['-',np.NaN,np.nan],'0')
data['GA sessions']         =data['GA sessions'].replace(['-',np.NaN,np.nan],'0')
data['SC Impressions']      =data['SC Impressions'].replace(['-',np.NaN,np.nan],'0')
```

```python
data['searchVolume MONTHLY']=data['searchVolume MONTHLY'].astype(int)
data['Transactions'] =data['Transactions'].astype(int)
data['SC Clicks']=data['SC Clicks'].astype(int)
data['GA sessions'] =data['GA sessions'].astype(int)
data['SC Impressions'] =data['SC Impressions'].astype(int)
```

```python
data.dtypes
```

```python
data['sorting_columns'] = ((data['Transactions_e']*data['Transactions'])+
                          (data['SC Clicks_e']*data['SC Clicks'])+
                          (data['GA sessions_e']*data['GA sessions'])+
                          (data['SC Impressions_e']*data['SC Impressions'])+
                          (data['searchVolume MONTHLY_e']*data['searchVolume MONTHLY']))
```

```python
data['Rank']= data['sorting_columns'].rank(ascending=False)
```
